# Pymaceuticals Clinical Trial Summary


### Trial Overview:
250 mice were individually treated with one of four drug variations: Capomulin, Infubinol, Ketapril, and Placebo over the course of 45 days. The data corresponding to the physiological responses shows that at the trial conclusion, Capomulin significantly outperforms the other three treatments.



### Tumor Volume
Capomulin effectively reduced the tumor volume by 19% while the remainder of treatments showed an increase within a range of 46% - 57%.  The respective changes in volume are as follows (per mm3):

Capomulin: -8.78   .   Infubinol: +20.8    .    Ketapril:  +25.7   .    Placebo:   +23.1

### Survival Rates

Of the 250 mice, those treated with Capomulin showed the greatest survival rates at roughly 87% survived, compared to 35-45% of the mice receiving alternative drugs. The respective survival counts  are as follows:

Capomulin: 21   .   Infubinol: 9    .    Ketapril:  11   .    Placebo:   11
